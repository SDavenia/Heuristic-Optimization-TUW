```python
import pandas as pd
import os
import re
```

```python
# Initializing lists to store data
instances = []
nodes = []
edges = []
runtimes_c = []
scores_c = []
runtimes_rc = []
scores_rc = []
scores_ls_m1 = []
scores_ls_s2 = []
scores_ls_jc = []
scores_grasp_jc = []
scores_grasp_s2 = []
scores_vnd = []
scores_gvns = []
```

```python
# Reading the text file and extracting results for construction heuristic.
with open('solutions/summary_test/summary_c.txt', 'r') as file:
    lines = file.readlines()

for i in range(0, len(lines), 4):
    # Extract the instance name
    match_name = re.search(r'\/([^\/]+?)_(?:n|m)_', lines[i])
    instances.append(match_name.group(1))
    # Extract the number of nodes
    match_n = re.search(r'_n_(\d+)', lines[i])
    nodes.append(match_n.group(1))
    # Extract the number of edges
    match_m = re.search(r'_m_(\d+)', lines[i])
    edges.append(match_m.group(1))


for i in range(1, len(lines), 4):
    match_number = re.search(r'(\d+\.\d{0,3})', lines[i])
    runtimes_c.append(float(match_number.group(1)))

for i in range(2, len(lines), 4):
    match_number = re.search(r'(\d+)', lines[i])
    scores_c.append(float(match_number.group(1)))
```

```python
with open('solutions/summary_test/summary_rc.txt', 'r') as file:
    file_content = file.read()

# Extracting numbers after "Score:" using regular expression
scores_rc = re.findall(r'Score:\s(\d+)', file_content)

# Converting scores to integers
scores_rc = list(map(int, scores_rc))

# Printing the array of scores
print(scores_rc)
```
