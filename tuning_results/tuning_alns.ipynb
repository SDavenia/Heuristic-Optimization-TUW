```python
import pandas as pd

# Specify the path to your Excel file
excel_file_path = 'tuning_results/tuning_alns.xlsx'

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(excel_file_path)


df = df.rename(columns={df.columns[0]: 'instance'})  # Replace 'old_column_name' with the actual name of the first column
df.head()
```

```python
df = df[['instance', 'join_rate', 'Mean final']]
```

```python
min_idx = df.groupby('instance')['Mean final'].idxmin()

# Extract the corresponding rows
result_df = df.loc[min_idx]

# Display the result
print(result_df)
```
